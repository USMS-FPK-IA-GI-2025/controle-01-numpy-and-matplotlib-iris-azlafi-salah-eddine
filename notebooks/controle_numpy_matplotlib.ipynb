# Partie A — Importation du Dataset, Arrays, Premières Manipulations

### A1. Importation du dataset
Importez NumPy et Matplotlib.  
Chargez le fichier `iris.csv` avec `np.genfromtxt` (`delimiter=','`, `skip_header=1`).  
Stockez les données dans un array nommé `data` et affichez sa `shape`.


```python
# Votre code ici
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt(
    'Iris.csv',
    delimiter=',',
    skip_header=1,
)
print(data.shape)
```

### A2. Séparation des features et des espèces
Séparez :
- les 4 premières colonnes numériques dans un array `features`
- la dernière colonne (espèces) dans un array `species`
Utilisez le slicing.


```python
# Votre code ici

features = data[:,1:5]
species = np.genfromtxt(
    'Iris.csv',
    delimiter=',',
    skip_header=1,
    usecols=5,
    dtype=str
)

print(features)
print(species)
```

### A3. Indexing
Affichez les 5 premières lignes de `features` en utilisant l’indexing.


```python
# Votre code ici

print(features[:5])
```

### A4. Extraction d’une colonne
Créez un array `sepal_length` contenant uniquement la première colonne de `features`.  
Affichez sa `size` et son `dtype`.


```python
# Votre code ici
sepal_length = features[:,0]
print(sepal_length.size)
print(sepal_length.dtype)
```

### A5. Slicing par espèce
Utilisez le slicing pour extraire les 50 premières lignes correspondant à l’espèce *Setosa*.  
Stockez-les dans `setosa_features`.


```python
# Votre code ici
setosa_features = features[:50 , :]

print(setosa_features.shape)
```

# Partie B — Statistiques Basiques et Opérations Mathématiques

### B6. Moyenne par colonne
Calculez la moyenne de chaque colonne de `features` avec `np.mean(axis=0)`.


```python
# Votre code ici

mean_features = np.mean(features,axis=0)

print(mean_features)
```

### B7. Min et Max
Trouvez la valeur minimale et maximale de la longueur des pétales.


```python
# Votre code ici
petal_length = features[:,2]
min_petal_length = np.min(petal_length)
max_petal_length = np.max(petal_length)

print(min_petal_length)
print(max_petal_length)
```

### B9. Normalisation
Normalisez les longueurs des pétales en les divisant par leur maximum.
Stockez le résultat dans `normalized_petals`.


```python
# Votre code ici
petal_length = features[:,2]
max_petal_length = np.max(petal_length)
print("max :",max_petal_length)
normalized_petals = petal_length/max_petal_length

print(normalized_petals)
```

### B10. Somme cumulée
Calculez la somme cumulée des largeurs des pétales pour Versicolor.
Affichez les 10 premières valeurs.


```python
# Votre code ici
versicolor_features = features[50:100,:]

petal_width_versicolor = versicolor_features[:,3]

cumsum_petal_width_versicolor = np.cumsum(petal_width_versicolor)

print(cumsum_petal_width_versicolor[:10])
```

### C13. Concaténation
Concaténez horizontalement les moyennes par colonne avec un array de zéros.


```python
# Votre code ici
mean_features = np.mean(features,axis=0)
mean_features = mean_features.reshape(1, 4)
zeros_array = np.zeros((1,4))
concatenated_array = np.concatenate((mean_features,zeros_array),axis=0)

print(concatenated_array)
```

Simulez des valeurs NaN et remplacez-les avec `np.nan_to_num`.

### E26. Line plot superposé
Cumsum des largeurs des pétales par espèce.


```python
# Votre code ici

setosa_features = features[:50,:]
versicolor_features = features[50:100,:]
virginica_features = features[100:150,:]

petal_width_setosa = setosa_features[:,3]
petal_width_versicolor = versicolor_features[:,3]
petal_width_virginica = virginica_features[:,3]

cumsum_petal_width_setosa = np.cumsum(petal_width_setosa)
cumsum_petal_width_versicolor = np.cumsum(petal_width_versicolor)
cumsum_petal_width_virginica = np.cumsum(petal_width_virginica)

plt.plot(cumsum_petal_width_setosa,color='red',label='Setosa')
plt.plot(cumsum_petal_width_versicolor,color='blue',label='Versicolor')
plt.plot(cumsum_petal_width_virginica,color='green',label='Virginica')
plt.xlabel("Index")
plt.ylabel("Cumsum")
plt.title("Cumsum des largeurs des pétales par espèce")
plt.legend()
```

### F27. Corrélations
Calculez la matrice de corrélation avec `np.corrcoef`.


```python
# Votre code ici

correlation_matrix = np.corrcoef(features,rowvar=False)

print(correlation_matrix)
```

### F28. Visualisation des corrélations
Visualisez les corrélations sous forme de bar chart ou scatter.


```python
# Votre code ici

plt.scatter(correlation_matrix[0,1],correlation_matrix[0,2],color='red', label='(0,1) vs (0,2)')
plt.scatter(correlation_matrix[0,3],correlation_matrix[1,0],color='blue', label='(0,3) vs (1,0)')
plt.scatter(correlation_matrix[1,2],correlation_matrix[1,3],color='green', label='(1,2) vs (1,3)')
plt.scatter(correlation_matrix[2,0],correlation_matrix[2,3],color='yellow', label='(2,0) vs (2,3)')
plt.scatter(correlation_matrix[3,0],correlation_matrix[3,1],color='purple', label='(3,0) vs (3,1)')
plt.xlabel("Correlation Value 1")
plt.ylabel("Correlation Value 2")
plt.title("Visualisation des corrélations (paires d'indices)")
plt.legend()
plt.grid(True)
plt.show()
```

### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
# Votre code ici

setosa_features = features[:50,:]
versicolor_features = features[50:100,:]
virginica_features = features[100:150,:]

petal_length_setosa = setosa_features[:,2]
petal_length_versicolor = versicolor_features[:,2]
petal_length_virginica = virginica_features[:,2]

variance_setosa = np.var(petal_length_setosa)
variance_versicolor = np.var(petal_length_versicolor)
variance_virginica = np.var(petal_length_virginica)

print("Variance Setosa :",variance_setosa)
print("Variance Versicolor :",variance_versicolor)
print("Variance Virginica :",variance_virginica)

plt.subplot(1,3,1)
plt.hist(petal_length_setosa,bins=10,edgecolor='black')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.title("Setosa")

plt.subplot(1,3,2)
plt.hist(petal_length_versicolor,bins=10,edgecolor='black')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.title("Versicolor")

plt.subplot(1,3,3)
plt.hist(petal_length_virginica,bins=10,edgecolor='black')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.title("Virginica")

```
